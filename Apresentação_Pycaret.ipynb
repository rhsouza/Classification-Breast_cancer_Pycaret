## Auto ML : Previsão de câncer de mama utilizando pycaret

### Objetivo: classificar se um tumor é benigno ou maligno.

*Dataset criado por Dr. William H. Wolberg, W. Nick Street, Olvi L. Mangasarian. November, 1995.*

*Saiba mais em https://goo.gl/U2Uwz2*

### Sobre o Dataset:

Este conjunto de dados contém informações sobre características de tumores de mama, calculadas a partir de uma imagem digitalizada. As características descrevem a célula presente nas imagens.
Estas características são calculadas para cada célula presente na imagem e, em seguida, as médias das características de todas as células de um tumor são calculadas. <br><br>

#### Classe do tumor (target) :

- Maligno (0)
- Benigno (1)

#### Características:

- Raio médio (mean radius)
- Textura média (mean texture)
- Perímetro médio (mean perimeter)
- Área média (mean area)
- Suavidade média (mean smoothness)
- Compacidade média (mean compactness)
- Concavidade média (mean concavity)
- Pontos côncavos médios (mean concave points)
- Simetria média (mean symmetry)
- Dimensão Fractal Média (mean fractal dimension)
- Erro de Raio (radius error)
- Erro de Textura (texture error)
- Erro de Perímetro (perimeter error)
- Erro de Área (area error)
- Erro de Suavidade (smoothness error)
- Erro de Compactação (compactness error)
- Erro de Concavidade (concavity error)
- Erro de Pontos Côncavos (concave points error)
- Erro de Simetria (symmetry error)
- Erro de Dimensão Fractal (fractal dimension error)
- Pior Raio (worst radius)
- Pior Textura (worst texture)
- Pior Perímetro (worst perimeter)
- Pior Área (worst area)
- Pior Suavidade (worst smoothness)
- Pior Compacidade (worst compactness)
- Pior Concavidade (worst concavity)
- Piores Pontos Côncavos (worst concave points)
- Pior Simetria (worst symmetry)
- Pior Dimensão Fractal (worst fractal dimension)

#### Detalhes:

- Médias (mean): Representam os valores médios de cada célula presente na imagem.
- Erros (error): Representam os erros padrão das medidas. Ex: O erro de raio (radius error) indica a variabilidade ou incerteza nas medidas do raio das células.

- Piores Valores (worst): Representam os piores valores (ou os maiores valores) encontrados entre as células presentes em uma imagem. Ex: O pior raio (worst radius) indica o maior valor de raio entre todas as células.

- Raio (radius): Refere-se à distância do centro até a borda da célula.

- Perímetro (perimeter): Refere-se ao comprimento da linha que delimita a célula, em outras palavras, é comprimento da linha da borda.

- Área (area): Refere-se à medida da superfície da célula de forma bidimensional.

- Textura (texture): Refere-se à variação nos tons de cinza das imagens.

- Suavidade (smoothness): Refere-se à variação local nos comprimentos dos raios da célula.

### Importando e visualizando dados

```python
from pycaret.datasets import get_data

datasets = get_data()
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
#import sweetviz as sv

#viz = sv.analyze(df)
#viz.show_notebook()
```

```python
from sklearn.model_selection import train_test_split

# Dividindo os dados em conjuntos de treinamento e teste
train, test = train_test_split(df, test_size=0.2, random_state=13)
print('base treino:', train.shape, '\nbase teste:',test.shape)
```

### Pycaret
*setup doc: https://pycaret.readthedocs.io/en/latest/api/classification.html#pycaret.classification.setup*

*cheat sheet: https://pycaret.gitbook.io/docs/learn-pycaret/cheat-sheet*


#### Configurando o treinamento

```python
import pycaret.classification as pc

# Configuração do treinamento
pc.setup(session_id = 13,
      data = train,
      target ='target',
      train_size = 0.7)
      #test_data = test
```

#### Comparando modelos

```python
# Comparando diferentes modelos de classificação
best_model = pc.compare_models(sort = 'F1') #(Possibilidade de inclusão de modelo específico na grid)
# best_model = pc.compare_models(include = ['lr', 'dt', 'lightgbm'])
# best_model = pc.compare_models(exclude = ['lr', 'dt', 'lightgbm'])
# best_model = pc.compare_models(n_select = 3)
# best_model = pc.compare_models(budget_time = 0.5)
# best_model = pc.compare_models(probability_threshold = 0.7)
# best_model = pc.compare_models(cross_validation=False)
```

```python
# Parâmetros do modelo
print(best_model)
```

#### Criando modelo

```python
# Criação do modelo escolhido
et = pc.create_model('et')
```

#### Mesclando modelo

```python
#https://pycaret.gitbook.io/docs/get-started/functions/optimize
# Possibilidade de mesclar modelos
rf = pc.create_model('rf')
lr = pc.create_model('lr')
blender = pc.blend_models([et, rf, lr])
```

#### Otimizando modelo

```python
# Otimização de modelo
pc.tune_model(et, optimize = 'F1')
```

#### Avaliando modelo

```python
# Análise do modelo criado
pc.evaluate_model(et)
```

#### Aplicando previsão

```python
# Inferência com modelo criado
pc.predict_model(et, probability_threshold = 0.5, raw_score = True)
```

#### Finalizando criação de modelo

```python
#Finalização do modelo, aqui ele recria o modelo incluindo a base teste separada no setup
pc.finalize_model(et)
```

#### Salvando e carregando modelo criado

```python
# Salvando modelo e pipeline
pc.save_model(et, 'et_pipeline')

# Carregando modelo e pipeline
model = pc.load_model('et_pipeline')
```

```python
# O probability_threshold é um corte para a probabilidade de 1, ou seja, caso não atenda para nenhuma classe, atribuirá 0.
predictions = pc.predict_model(model, test, raw_score = True, probability_threshold = 0.5)
```
